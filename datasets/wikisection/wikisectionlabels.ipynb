```python
def find_examples(json_path: Path):
    data = json.loads(json_path.read_text())

    for wiki_article in data:

        text = wiki_article["text"]
        annotations = wiki_article["annotations"]

        for annotation in annotations:
            section = text[annotation["begin"]:annotation["begin"] + annotation["length"]].strip()
            section = section.replace("\\n", " ") # remove newlines
            section = section.replace("\\t", " ") # remove tabs
            section = " ".join(section.split()) # remove multiple whitespaces

            label = annotation["sectionLabel"]
            assert label.startswith(f"{category}.")
            label = label[len(f"{category}."):]

            labels_dict[label].append(section)
```

```python
find_examples(valid_path)
find_examples(test_path)
find_examples(train_path)
```
